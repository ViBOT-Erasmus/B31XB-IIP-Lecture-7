## Lecture 6 - Morphological Image Processing

The needed libraries are imported in the cell below

```python
%matplotlib inline
%pprint off

# Import the library to show images, plot, etc.
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
# Import functionality for the color map
import matplotlib.cm as cm

# Other plotting library
import seaborn as sns
sns.set()

# Import the library to manage matrix and array
import numpy as np

# Import the library of linear algebra
import scipy as sp

# Import the image processing toolbox
### Module to read, write, etc.
from skimage import io
### Morphological module
from skimage import morphology

from skimage import img_as_bool
from skimage import img_as_ubyte
```

### Structuring elements

```python
se_diamond = morphology.diamond(2)
plt.figure(figsize=(5, 5))
sns.heatmap(se_diamond, annot=True, fmt='d', linewidths=.5)
plt.show()

se_disk = morphology.disk(2)
plt.figure(figsize=(5, 5))
sns.heatmap(se_disk, annot=True, fmt='d', linewidths=.5)
plt.show()

se_octagon = morphology.octagon(2, 3)
plt.figure(figsize=(5, 5))
sns.heatmap(se_octagon, annot=True, fmt='d', linewidths=.5)
plt.show()

se_square = morphology.square(4)
plt.figure(figsize=(5, 5))
sns.heatmap(se_square, annot=True, fmt='d', linewidths=.5)
plt.show()

se_star = morphology.star(2)
plt.figure(figsize=(5, 5))
sns.heatmap(se_star, annot=True, fmt='d', linewidths=.5)
plt.show()
```

### Erosion

```python
# Load the image
im = img_as_ubyte(io.imread('./images/FindingContours.png', as_grey=True))

# Show the image
io.imshow(im)

# Erode the image using different size of disk
disk_sz = [1, 10, 20]

for sz in disk_sz:
    er_im  = morphology.erosion(im, morphology.disk(sz))
    plt.figure()
    io.imshow(er_im)
```

### Dilation

```python
# Load the image
im = img_as_ubyte(io.imread('./images/FindingContours.png', as_grey=True))

# Show the image
io.imshow(im)

# Erode the image using different size of disk
disk_sz = [1, 10, 20]

for sz in disk_sz:
    er_im  = morphology.dilation(im, morphology.disk(sz))
    plt.figure()
    io.imshow(er_im)
```

```python
# Load the image
im = img_as_ubyte(io.imread('./images/FindingContours.png', as_grey=True))

# Show the image
io.imshow(im)

# Erode the image using different size of disk
disk_sz = [1, 10, 20]

for sz in disk_sz:
    er_im  = morphology.opening(im, morphology.disk(sz))
    plt.figure()
    io.imshow(er_im)
```
